```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import requests
from bs4 import BeautifulSoup
```

```python
#Bundesliga

df_bundesliga=pd.DataFrame()
for i in range(2014,2024):
        # URL of the webpage to scrape
        url = f"https://www.transfermarkt.com/bundesliga/torverteilungart/wettbewerb/L1/plus/?saison_id={i}"

        # Set headers to mimic a browser visit
        headers = {
            "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/85.0.4183.121 Safari/537.36"
        }

        # Send a GET request to the webpage
        response = requests.get(url, headers=headers)

        # Parse the content of the page with BeautifulSoup
        soup = BeautifulSoup(response.content, "html.parser")

        # Find the correct table (inspect the website to find the right table)
        table = soup.find("table", class_="items") # The class name was incorrect

        # Extract table headers
        headers = []
        for th in table.find_all("th"):
            headers.append(th.text.strip())

        # Extract table rows
        rows = []
        for row in table.find_all("tr")[1:]:  # Skip the header row
            cells = row.find_all("td")
            rows.append([cell.text.strip() for cell in cells])

        # Create a DataFrame from the extracted data
        df = pd.DataFrame(rows, columns=headers)
        df['Year']=i
        # Display the DataFrame
        df_bundesliga=pd.concat([df,df_bundesliga])
df_bundesliga=df_bundesliga[df_bundesliga['Club']!='Total']
df_bundesliga['League']='Bundesliga'
df_bundesliga['Matches']=34
df_bundesliga
```

```python
all_leagues=pd.concat([df_bundesliga,df_EPL,df_IT,df_laliga,df_L1,df_POG,df_PL2])
all_leagues
```

```python
all_leagues[['Matches','Penalty','Right-footed','Left-footed','Header','Freekick','Total']]=all_leagues[['Matches','Penalty','Right-footed','Left-footed','Header','Freekick','Total']].astype(int)
```

```python
all_leagues_19_23=all_leagues[all_leagues['Year']>=2019]
all_leagues_14_18=all_leagues[all_leagues['Year']<2019]
```

```python
AL19_23=all_leagues_19_23.groupby('League')[['Right-footed','Left-footed','Matches']].sum()
AL14_18=all_leagues_14_18.groupby('League')[['Right-footed','Left-footed','Matches']].sum()
AL19_23['perLF']=AL19_23['Left-footed']/AL19_23['Matches']
AL14_18['perLF']=AL14_18['Left-footed']/AL14_18['Matches']
AL19_23['perRF']=AL19_23['Right-footed']/AL19_23['Matches']
AL14_18['perRF']=AL14_18['Right-footed']/AL14_18['Matches']
```

```python
AL19_23
```

```python
AL14_18.reset_index(inplace=True)
AL19_23.reset_index(inplace=True)
```

```python
# Create a figure with 2 subplots (side by side)
fig, ax = plt.subplots(1, 2, figsize=(14, 6))  # Adjust figsize as needed

# Create the scatter plots
sns.scatterplot(data=AL14_18, x='perLF', y='perRF', hue='League', sizes=(300, 300), s=200, ax=ax[0])
sns.scatterplot(data=AL19_23, x='perLF', y='perRF', hue='League', sizes=(300, 300), s=200, ax=ax[1])

# Set titles and labels
ax[0].set_title('Right foot vs Left foot 2014-19')
ax[0].set_xlabel('Goals per match from Left Foot')
ax[0].set_ylabel('Goals per match from Right Foot')
ax[1].set_title('Right foot vs Left foot 2019-23')
ax[1].set_xlabel('Goals per match from Left Foot')
ax[1].set_ylabel('Goals per match from Right Foot')

# Set grid and legend
ax[0].grid(True)
ax[1].grid(True)
ax[0].legend(title='League', bbox_to_anchor=(1.05, 1), loc='upper left')
ax[1].legend(title='League', bbox_to_anchor=(1.05, 1), loc='upper left')


# Set the same x and y limits for both subplots
ax[0].set_xlim(0.2,0.5)
ax[0].set_ylim(0.4, 0.75)
ax[1].set_xlim(0.2,0.5)
ax[1].set_ylim(0.4, 0.75)

# Adjust layout to fit the plots and labels
plt.tight_layout()

# Show the plot
plt.savefig('RFvsLF.png')

plt.show()
```
